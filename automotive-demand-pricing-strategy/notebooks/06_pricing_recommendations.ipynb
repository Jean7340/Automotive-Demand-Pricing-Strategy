## Business Question

Given the demand model, price elasticity estimates, and competitive environment, what pricing strategy should management adopt?

This notebook summarizes the analytical findings and evaluates alternative pricing strategies from a business perspective.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cars = pd.read_csv("../data/cars.csv", encoding="latin1")
cars.columns = cars.columns.str.strip()
```

```python
strategies = pd.DataFrame({
    "Strategy":[
        "Increase Price (+5%)",
        "Maintain Current Price",
        "Reduce Price (-5%)"
    ],

    "Price Change":[
        0.05,
        0.00,
        -0.05
    ]
})

strategies
```

```python
elasticity = -1.37

strategies["Demand Change"] = elasticity * strategies["Price Change"]

strategies["Revenue Change"] = (
    (1+strategies["Price Change"])
    *
    (1+strategies["Demand Change"])
    -1
)

strategies
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(
    strategies["Strategy"],
    strategies["Revenue Change"]*100
)

plt.axhline(0,color="gray")

plt.ylabel("Estimated Revenue Change (%)")

plt.title("Estimated Revenue Impact of Pricing Strategies")

plt.show()
```

```python
recommendation = pd.DataFrame({

    "Strategy":[
        "Increase Price (+5%)",
        "Maintain Price",
        "Reduce Price (-5%)"
    ],

    "Expected Revenue":[
        "Decrease",
        "Stable",
        "Increase"
    ],

    "Risk":[
        "High",
        "Low",
        "Medium"
    ],

    "Recommendation":[
        "Avoid",
        "Acceptable",
        "Preferred"
    ]
})

recommendation
```

```python
strategy_matrix = pd.DataFrame({

    "Scenario":[
        "Elastic Demand",
        "Moderately Elastic",
        "Strong Competition",
        "Weak Competition"
    ],

    "Pricing Decision":[
        "Reduce Price",
        "Maintain Price",
        "Maintain Price",
        "Increase Price Selectively"
    ]
})

strategy_matrix
```

### Recommendation 1

Avoid broad portfolio-wide price increases because the estimated demand elasticity suggests that demand is price-sensitive.

### Recommendation 2

Maintain current prices in highly competitive market segments unless competitors initiate price increases.

### Recommendation 3

Consider selective price increases for differentiated products or premium brands where competitive pressure is weaker.

### Recommendation 4

Future work should estimate segment-level elasticities and competitor reactions to support dynamic pricing decisions.

## Conclusion

This project combines demand estimation, price elasticity analysis, and competitive dynamics to support pricing decisions in the European automotive market.

Key findings indicate that:

- Vehicle demand is price-sensitive.
- Competitor prices influence substitution behavior.
- Competitive intensity affects pricing power.
- Broad price increases are unlikely to maximize revenue.

Instead, pricing decisions should be differentiated by market segment, product positioning, and competitive environment.
